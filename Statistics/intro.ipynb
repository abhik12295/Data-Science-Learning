```python
'''
Statistics:
    - is a field of mathematics that deals with the collection, analysis, interpretation, presentation, 
        and organization of data.
    - helps us for decision making in the presence of uncertainty.
    - through charts, graphs, and tables, we can visualize the data.
    - understanding the data is important for making informed decisions.
    - Application of statistics:
        - Business: sales forecasting, quality control, market research
        - Medicine: clinical trials, epidemiology
        - Social Sciences: surveys, experiments
        - Sports: performance analysis, player statistics
    - Types of statistics:
        - Descriptive statistics: summarize and describe the main features of a dataset
            - Measures of central tendency: mean, median, mode
            - Measures of variability(Dispersion): range, variance, standard deviation
            - Measures of position: percentiles, quartiles
        - Inferential statistics: make inferences about a population based on a sample -> z-tests, t-tests, ANOVA
            - Hypothesis testing
            - Confidence intervals
            - Regression analysis
        - Bayesian statistics: incorporates prior knowledge and beliefs into the analysis
        - Non-parametric statistics: does not assume a specific distribution for the data
    - Data visualization: graphical representation of data
        - Bar charts
        - Histograms
        - Box plots
        - Scatter plots
        - Heatmaps
    - Data cleaning: process of identifying and correcting errors in the data
    - Data transformation: process of converting data from one format to another
    - Data normalization: process of scaling data to a specific range
    - Data standardization: process of transforming data to have a mean of 0 and a standard deviation of 1
    - Data imputation: process of filling in missing values in a dataset
    - Data sampling: process of selecting a subset of data from a larger dataset
    - Data mining: process of discovering patterns and knowledge from large amounts of data
    - Data analysis: process of inspecting, cleansing, transforming, and modeling data

'''
```

```python
'''
Population vs Sample:
    - Population: entire group of individuals or items that we want to study
    - Sample: subset of the population that we actually observe
    - Sampling methods:
        - Random sampling: each member of the population has an equal chance of being selected
        - Stratified sampling: population is divided into subgroups, and a random sample is taken from each subgroup
        - Systematic sampling: every nth member of the population is selected
        - Convenience sampling: sample is taken from a group that is easily accessible

'''
```

```python

'''
Descriptive statistics:
    - Measures of central tendency: mean, median, mode
        - Mean: average of a dataset
            formula:  u = Σi=1^n xi / n
            where u is the mean, xi is the ith value in the dataset, and n is the number of values in the dataset
            - sample formula: x̄ = Σi=1^n xi / n
            where x̄ is the sample mean, xi is the ith value in the dataset, and n is the number of values in the dataset
        
        - Median: middle value of a dataset when arranged in ascending or descending order
            - If n is odd: median = x((n+1)/2)
            - If n is even: median = (x(n/2) + x((n/2)+1)) / 2
            where x is the ith value in the dataset, and n is the number of values in the dataset
            - example: 1, 2, 3, 4, 5 -> median = 3
            - example: 1, 2, 3, 4 -> median = (2 + 3) / 2 = 2.5
            - median is less affected by outliers than the mean
            - In a skewed distribution, the median is a better measure of central tendency than the mean

        - Mode: most frequently occurring value in a dataset
            - A dataset can have more than one mode (bimodal, multimodal)
            - example: 1, 2, 2, 3, 4 -> mode = 2
            - example: 1, 2, 3, 4 -> no mode
            - If no value repeats, the dataset has no mode
            - Mode is useful for categorical data



    - Measures of variability(Dispersion): range, variance, standard deviation
        - Range: difference between the maximum and minimum values in a dataset
            formula: R = x(max) - x(min)
            where R is the range, x(max) is the maximum value in the dataset, and x(min) is the minimum value in the dataset
            - Range is sensitive to outliers
            - example: 1, 2, 3, 4, 5 -> range = 5 - 1 = 4
            - disadvantage: does not provide information about the distribution of the data


        - Variance: measure of how far each number in the dataset is from the mean
            - population formula: σ^2 = Σi=1^n (xi - u)^2 / n
            where σ^2 is the variance, xi is the ith value in the dataset, u is the mean, and n is the number of values in the dataset
            - Variance is sensitive to outliers
            - example: 1, 2, 3, 4, 5 -> variance = ((1-3)^2 + (2-3)^2 + (3-3)^2 + (4-3)^2 + (5-3)^2) / 5 = 2
            - disadvantage: variance is in squared units of the original data
            - sample formula: s^2 = Σi=1^n (xi - x̄)^2 / (n-1)
            where s^2 is the sample variance, xi is the ith value in the dataset, x̄ is the sample mean, and n is the number of values in the dataset

        - Standard deviation: square root of the variance
            formula: σ = √(Σi=1^n (xi - u)^2 / n)
            where σ is the standard deviation, xi is the ith value in the dataset, u is the mean, and n is the number of values in the dataset
            - Standard deviation is sensitive to outliers
            - example: 1, 2, 3, 4, 5 -> standard deviation = √(2) = 1.414
            - advantage: standard deviation is in the same units as the original data
            - disadvantage: standard deviation is sensitive to outliers


    - Measures of position: percentiles, quartiles
        - Percentiles: value below which a percentage of the data falls
        - Quartiles: divide the data into four equal parts

'''
```
