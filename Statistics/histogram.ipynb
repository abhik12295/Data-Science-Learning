```python
'''
Histogram:
    defintion: A histogram is a graphical representation of the distribution of numerical data.
    It is an estimate of the probability distribution of a continuous variable.
    It is a type of bar plot where the x-axis represents the bins and the y-axis represents the frequency of data points in each bin.
    example: A histogram of the heights of a group of people.
    It shows how many people fall into each height range (bin).
    It is used to visualize the distribution of a dataset and to identify patterns, trends, and outliers.
    heights = [150, 160, 170, 180, 190, 200]
    bins_range = [100-200, 200-300]
    histogram = Histogram(heights, bins_range)
    histogram.plot()
    
'''
```

```python
heights_range = [120, 150, 160, 170, 180, 190, 200]
height_people = [128,135,155,160,165,165,167,170,175,175,180,185,185,189,190,195,195]

import matplotlib.pyplot as plt
import numpy as np
class Histogram:
    def __init__(self, data, bins_range):
        self.data = data
        self.bins_range = bins_range

    def plot(self):
        plt.hist(self.data, bins=self.bins_range, edgecolor='white', alpha=0.7)
        plt.xlabel('Height (cm)')
        plt.ylabel('Frequency')
        plt.title('Histogram of Heights')
        plt.show()
# Create a histogram object
histogram = Histogram(heights_range, height_people)
# Plot the histogram
# The code above creates a histogram of the heights of a group of people.
# It shows how many people fall into each height range (bin).
histogram.plot()
```

```python
heights_range = [120, 150, 160, 170, 180, 190, 200]
height_people = [128,135,155,160,165,165,167,170,175,175,180,185,185,189,190,195,195]

import matplotlib.pyplot as plt
import numpy as np

#kernel density estimation - 
# Kernel Density Estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable.
# It is used to smooth out the noise in the data and to visualize the distribution of the data.
# It is a way to estimate the probability density function of a random variable.

plt.hist(height_people, bins=heights_range, density=True, alpha=0.5, color='g')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Heights')
plt.show()
```
