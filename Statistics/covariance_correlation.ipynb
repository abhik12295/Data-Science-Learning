```python
'''
Covariance:
    definition: Covariance is a measure of how much two random variables change together
    If the variables tend to increase together, the covariance is positive. 
    If one variable tends to increase when the other decreases, the covariance is negative.
    If the variables are independent, the covariance is zero.
    Covariance is not normalized, so it can take any value from -∞ to +∞.
    Cov(X, Y) = 1/(n-1) * Σ (xi - x̄)(yi - ȳ)
    where n is the number of samples, xi and yi are the values of the random variables X and Y, 
    and x̄ and ȳ are the means of X and Y.
    
    Cov(X, Y) = E[(X - E[X])(Y - E[Y])]
    Cov(X, Y) = E[XY] - E[X]E[Y]
    Cov(X, X) = Var(X)
    Cov(X, Y) = 0 if X and Y are independent
    Cov(X, Y) = Cov(Y, X)
    Cov(X + Y, Z) = Cov(X, Z) + Cov(Y, Z)
    Cov(aX + bY, Z) = aCov(X, Z) + bCov(Y, Z)
    Cov(X, Y + Z) = Cov(X, Y) + Cov(X, Z)
    Cov(X, aY + bZ) = aCov(X, Y) + bCov(X, Z)

    Limit range:
        interval: (-∞, +∞)
        domain: (-∞, +∞)
    
'''

```

```python
'''
Correlation and Covariance are both measures of the relationship between two random variables.
Covariance is a measure of how much two random variables change together, while correlation is a measure of the strength and direction of the linear relationship between two random variables.
Correlation is a standardized version of covariance, which means it is scaled to be between -1 and 1.
Covariance can take any value, while correlation is always between -1 and 1.
'''
```

```python
'''
Covariance matrix is a square matrix that contains the covariances between 
all pairs of random variables in a multivariate distribution.
Covariance matrix is a generalization of the variance to multiple dimensions.
Covariance matrix is used to describe the shape of a multivariate distribution.

'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data: Hours Studied vs Exam Score
hours_studied = np.array([1, 2, 3, 4, 5, 6, 7])
exam_score = np.array([50, 55, 65, 70, 75, 85, 90])

# Create DataFrame
df = pd.DataFrame({
    'Hours_Studied': hours_studied,
    'Exam_Score': exam_score
})
```

```python
# Covariance matrix
cov_matrix = np.cov(df['Hours_Studied'], df['Exam_Score'])
print("Covariance matrix:\n", cov_matrix)
print("Covariance value:", cov_matrix[0, 1])

# Correlation matrix
corr_matrix = np.corrcoef(df['Hours_Studied'], df['Exam_Score'])
print("\nCorrelation matrix:\n", corr_matrix)
print("Correlation coefficient:", corr_matrix[0, 1])
```

```python
# Scatter plot with regression line
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.regplot(x='Hours_Studied', y='Exam_Score', data=df, marker='o', color='green')
plt.title('Scatter Plot: Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# example to calclate covariance and correlation
# Sample data
data = {
    'X': [2, 3, 4, 5, 6],
    'Y': [50, 60, 70, 80, 90]
}
df = pd.DataFrame(data)
# Calculate covariance
#using formula - calculate mean, subtract mean from each value, multiply the results, sum them up and divide by n-1
mean_x = df['X'].mean()
print("Mean of X:", mean_x)
mean_y = df['Y'].mean()
print("Mean of Y:", mean_y)
cov_xy = ((df['X'] - mean_x) * (df['Y'] - mean_y)).sum() / (len(df) - 1)
print("Covariance (X, Y):", cov_xy)
# Calculate correlation
#using formula - calculate covariance and calculate standard deviation of each variable
std_x = df['X'].std(ddof=0)  # population standard deviation
std_y = df['Y'].std(ddof=0)  # population standard deviation
print("Standard deviation of X:", std_x)
print("Standard deviation of Y:", std_y)
corr_xy = cov_xy / (std_x * std_y)
print("Correlation (X, Y):", corr_xy)

```

```python
'''

Correlation:
    definition: Correlation is a measure of the strength and direction of the linear relationship between two random variables
    Correlation is normalized  version of covariance, so it takes values from -1 to 1.
    If the variables are perfectly positively correlated, the correlation is 1.
    If the variables are perfectly negatively correlated, the correlation is -1.
    If the variables are independent, the correlation is 0.
    
    Corr(X, Y) = Cov(X, Y) / (σX * σY)
    where σX and σY are the standard deviations of X and Y.
    formula: for standard deviation:
    σX = sqrt(Var(X)) = sqrt(E[X^2] - E[X]^2)
    σY = sqrt(Var(Y)) = sqrt(E[Y^2] - E[Y]^2)
    where E[X] and E[Y] are the means of X and Y.
    
    Corr(X, Y) = 0 if X and Y are independent
    Corr(X, Y) = 1 if X and Y are perfectly positively correlated
    Corr(X, Y) = -1 if X and Y are perfectly negatively correlated

    limit range:
        interval: (-1, 1)
        domain: (-1, 1)
'''
```

```python
'''
Types of correlation:
    - Pearson correlation: measures the linear relationship between two variables
    - Spearman correlation: measures the monotonic relationship between two variables 
        (monotonic means that the relationship is either increasing or decreasing, but not necessarily linear)
    - Kendall correlation: measures the ordinal relationship between two variables
    - Point-biserial correlation: measures the relationship between a continuous variable and a binary variable
'''
```

```python
'''
Pearson correlation:
    definition: Pearson correlation is a measure of the linear relationship between two variables
    It is the most common type of correlation used in statistics.
    It is sensitive to outliers, so it may not be appropriate for data with extreme values.
    It assumes that the data is normally distributed and that the relationship between the variables is linear.
    
    formula: 
    Corr(X, Y) = Cov(X, Y) / (σX * σY)
    where σX and σY are the standard deviations of X and Y.
    Standard deviation is calculated as basically the square root of the variance.
    Variance is calculated as the average of the squared differences from the mean.

    Limit range:
        interval: (-1, 1)
        domain: (-1, 1)
        which means values closer to 1 is a strong positive correlation,
        while values closer to -1 indicate a strong negative correlation.
    Value = 0 indicates no correlation.
    Value = 1 indicates perfect positive correlation.
    Value = -1 indicates perfect negative correlation.
    Value <1 and >-1 indicates a weak correlation.
    Example: 
        If the Pearson correlation coefficient between two variables is 0.8, it means that there is a strong positive linear relationship between the two variables.
        If the Pearson correlation coefficient between two variables is -0.8, it means that there is a strong negative linear relationship between the two variables.
'''
```

```python
'''
Spearman correlation:
    definition: Spearman correlation is a measure of the monotonic relationship between two variables.
    (monotonic means that the relationship is either increasing or decreasing, but not necessarily linear)
    It is a non-parametric measure, which means it does not assume that the data is normally distributed.
    It is based on the ranks of the data rather than the actual values, so it is less sensitive to outliers.
    It can be used with ordinal data or continuous data that does not meet the assumptions of parametric tests.
    
    formula:
    Corr(X, Y) = 1 - (6 * Σ d_i^2) / (n * (n^2 - 1))
    where d_i is the difference between the ranks of the two variables for each observation,
    n is the number of observations.

    rs = Cov(Rank(X), Rank(Y)) / (σRank(X) * σRank(Y))
    where σRank(X) and σRank(Y) are the standard deviations of the ranks of X and Y.

    Example to calclate rank:
    x = [3, 1, 2]
    y = [30, 10, 20]
    R(x) = [2, 0, 1]  # ranks of x
    R(y) = [2, 0, 1]  # ranks of y
    Corr(Rank(X), Rank(Y)) = Cov(R(x), R(y)) / (σRank(R(x)) * σRank(R(y)))
    

    Limit range:
        interval: (-1, 1)
        domain: (-1, 1)
        which means values closer to 1 is a strong positive correlation,
        while values closer to -1 indicate a strong negative correlation.

    Value = 0 indicates no correlation.
    Value = 1 indicates perfect positive correlation.
    Value = -1 indicates perfect negative correlation.
    Value>0 indicates positive correlation.
    Value<0 indicates negative correlation.
'''
```

```python
# Example to calculate Spearman correlation
import math
def spearman_correlation(x, y):
    # Calculate ranks
    rank_x = pd.Series(x).rank()
    rank_y = pd.Series(y).rank()
    
    # Calculate covariance of ranks
    cov_rank = np.cov(rank_x, rank_y)[0, 1]
    
    # Calculate standard deviations of ranks
    std_rank_x = rank_x.std(ddof=0)
    std_rank_y = rank_y.std(ddof=0)
    
    # Calculate Spearman correlation
    rs = cov_rank / (std_rank_x * std_rank_y)
    return rs
# Sample data
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]
# Calculate Spearman correlation
spearman_corr = spearman_correlation(x, y)
print("Spearman correlation:", round(spearman_corr,2))
```
