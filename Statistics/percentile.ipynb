```python
'''
Percentile and Quantile:
    - Percentile: A value below which a given percentage of observations fall.
    - Quantile: A value that divides the data into equal-sized intervals.

Percentiles are often used in statistics to understand the distribution of data.
    - For example, the 50th percentile (or median) is the value below which 50% of the data falls.
    - The 25th percentile (or first quartile) is the value below which 25% of the data falls.
    - The 75th percentile (or third quartile) is the value below which 75% of the data falls.

Quantiles are used to divide the data into equal-sized intervals.
    - For example, the 0.25 quantile (or first quartile) is the value below which 25% of the data falls.
    - The 0.5 quantile (or median) is the value below which 50% of the data falls.
    - The 0.75 quantile (or third quartile) is the value below which 75% of the data falls.
    - The 0.1 quantile is the value below which 10% of the data falls.

Difference:
    - Percentiles are specific values that divide the data into intervals based on percentages.
    - Quantiles are values that divide the data into equal-sized intervals.
    - Percentiles are often used to understand the distribution of data, while quantiles are used to divide the data into equal-sized intervals.

    Example:
    - percentiles: 25th, 50th, 75th percentiles
    - quantiles: 0.25, 0.5, 0.75 quantiles
'''
```

```python
from numpy import percentile
import matplotlib.pyplot as plt
import numpy as np

# Data
numbers = [3, 7, 8, 5, 12, 14, 21, 13, 18]

# Calculate specific percentiles
percentile_25 = percentile(numbers, 25)
percentile_50 = percentile(numbers, 50)
percentile_75 = percentile(numbers, 75)

print(f"25th Percentile: {percentile_25}")
print(f"50th Percentile (Median): {percentile_50}")
print(f"75th Percentile: {percentile_75}")

# Plot: Percentile (x-axis) vs Value (y-axis)
percentiles = np.arange(0, 101, 1)
values = [percentile(numbers, p) for p in percentiles]

plt.figure(figsize=(10, 6))
plt.plot(percentiles, values, color='blue', label='Percentile Curve')
plt.axhline(percentile_25, color='red', linestyle='--', label='25th Percentile')
plt.axhline(percentile_50, color='green', linestyle='--', label='50th Percentile (Median)')
plt.axhline(percentile_75, color='orange', linestyle='--', label='75th Percentile')

plt.xlabel('Percentile (%)')
plt.ylabel('Value')
plt.title('Percentile Plot of Numbers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = [3, 7, 8, 5, 12, 14, 21, 13, 18]
data = np.sort(data)  # sort data for quantile visualization

# Quantiles to compute (from 0 to 1)
quantiles = np.linspace(0, 1, 101)

# Get quantile values
quantile_values = np.quantile(data, quantiles)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(quantiles, quantile_values, color='blue', marker='o', markersize=3)
plt.title('Quantile Plot (Inverse CDF)')
plt.xlabel('Quantile (0 to 1)')
plt.ylabel('Data Value')
plt.grid(True)

# Optional: Show specific quantile lines
for q in [0.25, 0.5, 0.75]:
    plt.axvline(x=q, color='gray', linestyle='--')
    plt.text(q, np.quantile(data, q), f'{int(q*100)}%', verticalalignment='bottom')

plt.show()
```
