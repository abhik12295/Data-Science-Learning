```python
'''Data Visualization with Seaborn
based on matplotlib -> high level interface for drawing attractive and informative stastical graphics
-> creates complex visualizations with just few lines of code
'''
```

```python
import seaborn as sns
```

```python
# Basic plotting with seaborn
tips = sns.load_dataset('tips')
tips.head()
```

```python
# create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Tip Plot 1")
```

```python
# Line plot
sns.lineplot(x = 'size',y ='total_bill',data=tips)
plt.title("Line Plot")
plt.show()
```

```python
# Categorcal Plots
# Bar Plot
sns.barplot(x = 'day',y='total_bill', data = tips)
plt.title('Bar Plot of Total Bill by Day')
```

```python
# box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# violin plot = how distribution of data is wrt total bill device
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
#kde True is line curve on top
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# Pairplot = relationship b/w all variables to diff diff variables
# need numerical values (doesn't consider categorical variables)
sns.pairplot(tips)
```

```python
# Heatmap = correlation one variable to another variable
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data_sample.csv', encoding='ISO-8859-1')
sales_df.head()
sales_df.columns
```

```python
sales_df.head()
```

```python
#plot  total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='TERRITORY',y="MSRP", data=sales_df,estimator=sum)
plt.title('Total Sales by TERRITORY')
plt.xlabel('TERRITORY')
plt.ylabel('Total Sales')
plt.show()
```

```python
#plot  total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='COUNTRY',y="MSRP", data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```
