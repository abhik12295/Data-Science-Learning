```python
'''
Histogram:
    defintion: A histogram is a graphical representation of the distribution of numerical data.
    It is an estimate of the probability distribution of a continuous variable.
    It is a type of bar plot where the x-axis represents the bins and the y-axis represents the frequency of data points in each bin.
    example: A histogram of the heights of a group of people.
    It shows how many people fall into each height range (bin).
    It is used to visualize the distribution of a dataset and to identify patterns, trends, and outliers.
    heights = [150, 160, 170, 180, 190, 200]
    bins_range = [100-200, 200-300]
    histogram = Histogram(heights, bins_range)
    histogram.plot()
    
'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

heights_range = [120, 150, 160, 170, 180, 190, 200]
height_people = [128, 135, 155, 160, 165, 165, 167, 170, 175, 175, 180, 185, 185, 189, 190, 195, 195]

class Histogram:
    def __init__(self, data, bins_range):
        self.data = data
        self.bins_range = bins_range

    def plot(self):
        plt.hist(self.data, bins=self.bins_range, edgecolor='white', alpha=0.7)
        plt.xlabel('Height (cm)')
        plt.ylabel('Frequency')
        plt.title('Histogram of Heights')
        plt.grid(True)
        plt.show()

# Correct the order: data first, then bins
histogram = Histogram(height_people, heights_range)

# Plot the histogram
histogram.plot()
```

```python
'''
Kernel Density Estimation (KDE):
    definition: Kernel Density Estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable.
    It is used to smooth out the noise in the data and to visualize the underlying distribution.
    KDE is often used as an alternative to histograms for visualizing the distribution of data.
    example: A KDE plot of the heights of a group of people.
    It shows the estimated probability density function of the heights.
    heights = [150, 160, 170, 180, 190, 200]
    kde = KernelDensity(heights)
    kde.plot()

Probability Density Function (PDF):
    definition: A Probability Density Function (PDF) is a function that describes the likelihood of a random variable taking on a particular value.
    It is used to model the distribution of continuous random variables.
    The area under the PDF curve represents the probability of the random variable falling within a certain range.
    example: A PDF of the heights of a group of people.
    It shows the likelihood of different heights occurring in the population.
    heights = [150, 160, 170, 180, 190, 200]
    pdf = ProbabilityDensityFunction(heights)
    pdf.plot()
'''
```

```python
heights_range = [120, 150, 160, 170, 180, 190, 200]
height_people = [128,135,155,160,165,165,167,170,175,175,180,185,185,189,190,195,195]


#kernel density estimation - 
# Kernel Density Estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable.
# It is used to smooth out the noise in the data and to visualize the distribution of the data.
# It is a way to estimate the probability density function of a random variable.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Data
height_people = [128, 135, 155, 160, 165, 165, 167, 170, 175, 175,
                 180, 185, 185, 189, 190, 195, 195]

# Reshape data for sklearn (expects 2D)
data = np.array(height_people).reshape(-1, 1)

# Fit KDE model
kde = KernelDensity(kernel='gaussian', bandwidth=2.0).fit(data)

# Define the range to evaluate density
x_vals = np.linspace(120, 200, 1000).reshape(-1, 1)

# Score_samples returns the log density, so we exponentiate it
log_dens = kde.score_samples(x_vals)
density = np.exp(log_dens)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, density, label='KDE (Gaussian, bw=2.0)')
plt.hist(height_people, bins=15, density=True, alpha=0.3, color='gray', label='Histogram')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.title('Kernel Density Estimate of Heights')
plt.legend()
plt.grid(True)
plt.show()

```
