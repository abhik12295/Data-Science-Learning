```python
from sklearn.datasets import fetch_california_housing
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cali_data = fetch_california_housing()
cali_data
```

```python
df = fetch_california_housing(as_frame=True).frame
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Price'] = cali_data.target
```

```python
df.drop(columns=['MedHouseVal'], inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
# Visualizing the data
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
df.head()
```

```python
# independent variables and dependent variable
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.head(1)
```

```python
y.head(1)
```

```python
# split test and train data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### Model Training and Evaluation

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model
```

```python
linear_model.fit(X_train, y_train)
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

```python
# Predicting the test set results
y_pred = linear_model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluation of the model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
print(f'Mean Absolute Error: {mae}')
```

```python
# adjsted r2 score
def adjusted_r2_score(r2, n, k):
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)
n = X_test.shape[0]  # number of observations
k = X_test.shape[1]  # number of features
adj_r2 = adjusted_r2_score(r2, n, k)
print(f'Adjusted R^2 Score: {adj_r2}')
```

### Assumptions

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
```

```python
residual = y_test - y_pred
residual
```

```python
# displaying the residuals
sns.displot(residual, kind='kde')
```

```python
# Sctter plot with prediction and residuals
plt.scatter(y_pred, residual)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Prices')
plt.show()
```

```python
'''
Pickeling the model:
Defintion: Pickling is the process of converting a Python object into a byte stream, 
which can then be saved to a file or transmitted over a network. 
This allows you to save the state of an object and later recreate it.
To pickle the model, you can use the `pickle` module in Python. Here's how you can do it:
'''
```

```python
import pickle
# write the model to a file
with open('linear_model.pkl', 'wb') as file:
    pickle.dump(linear_model, file)
```

```python
# read the model from a file
with open('linear_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
# Now you can use the loaded_model to make predictions
loaded_y_pred = loaded_model.predict(X_test)
print(loaded_y_pred)
```
